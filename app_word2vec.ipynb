```python
import jieba
import numpy as np 
import pandas as pd 

dfapp = pd.read_csv('zz_apps_info_22k',sep = '\001',encoding = 'utf-8')
app_name = dfapp['app_name'].values.tolist()

jieba.load_userdict(app_name)

app_info = dfapp['app_info'].values.tolist()

app_info_segs = [jieba.lcut(x) for x in app_info]
```

```python
import pkuseg

seg = pkuseg.pkuseg() 

app_info_segs = [seg.cut(x) for x in app_info]
```

```python
import gensim
import logging
from gensim import corpora, models, similarities

#logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
dictionary = corpora.Dictionary(app_info_segs)
dictionary.save('app_info.dict')  # store the dictionary

# 建立完整字典
dictionary.token2id
```

```python
len(dictionary)
```

```python
# 重点词典，去除文档频率2次以下和0.8以上的词
dictionary.filter_extremes(no_below=2,no_above=0.9,keep_n = 100000,keep_tokens = None)
dictionary.compactify()
#dictionary.token2id
```

```python
import re
with open('appwordvec.txt','r',encoding = 'utf-8') as fin,\
open('wordvec.txt','w',encoding = 'utf-8') as fout:
    for line in fin.readlines():
        if re.match('^[\d\,\.\。\；\！\、\*\?\;\：\【\】\，\!\:\ ]*$',line[0]) or line[0]=='\t':
            continue
        segs = line.rstrip().split(' ')
        if len(segs) !=101:
            continue
        else:
            fout.write('\t'.join(segs) + '\n')
```

### 计算词聚类

```python
import numpy as np 
import pandas as pd 

df = pd.read_csv('wordvec.txt',sep = '\t',header = None, encoding = 'utf-8',quoting=3)
```

```python
df.columns = ['word'] + ['vec_'+str(i) for i in range(1,101)]
```

```python
X = df.iloc[:,1:].values
```

```python
X.shape
```

```python
from sklearn.cluster import KMeans
```

```python
clt = KMeans(n_clusters = 500)
```

```python
word_clusters = clt.fit_predict(X)
```

```python
len(word_clusters)
```

```python
word_clusters
```

```python
dfcluster = df[['word']].copy()
```

```python
dfcluster['word_cluster'] = word_clusters.tolist()
```

```python
dfgroup = dfcluster.groupby('word_cluster').agg({'word':lambda l:','.join([x for x in l])})
```

```python
dfgroup['cluster'] = ['cluster_'+str(i) for i in range(len(dfgroup))]
```

```python
cluster = dfgroup.pop('cluster')
```

```python
dfgroup.insert(0,'cluster',cluster )
```

```python
dfgroup.to_csv('word_cluster.csv',sep = '\t',encoding = 'utf-8',index = None) 
```
